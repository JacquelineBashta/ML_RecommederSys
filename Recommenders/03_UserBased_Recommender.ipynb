# UserBased Recommender

## create Cross Table

```python
cross_table = pd.pivot_table(data=ratings_df, values="rating",columns="movieId",index="userId")
cross_table
```

## Replace NaN with = ( to use Cosine similarities )
thats not optimum to add impute with zero as it drag all the predicted weights to lower values...
TODO: can try to impute the values with something different like medien

```python
cross_table = cross_table.fillna(0)
```

```python
cross_table
```

## Calc Similarities between Users

```python
from sklearn.metrics.pairwise import cosine_similarity
usrs_similarities = pd.DataFrame(cosine_similarity(cross_table),
                                columns=cross_table.index, 
                                index=cross_table.index
                                )
```

```python
usrs_similarities
```

## calculate the weight from similarties (per user)

```python
chosen_user = 1
user_weight_col = usrs_similarities.query("userId!=@chosen_user")[chosen_user]
weights = user_weight_col/sum(user_weight_col)
weights
```

```python
# as test sum of weights should = 1
weights.sum()
```

## find items (movies) that chosen_user has not rated

```python
cross_table
```

```python
not_rated_items = cross_table.loc[cross_table.index!=chosen_user , cross_table.loc[chosen_user,:]== 0 ]
not_rated_items.T
```

## compute predicted rating that user would give (based on weighted similarity wwith other users)

```python
predicted_rates = pd.DataFrame(not_rated_items.T.dot(weights),columns=["predicted_rates"])
predicted_rates
```

## use predicted rates to compute top n items for chosen_user

```python
n_top = 10
recommended_items = predicted_rates.merge(movies_df, on="movieId")[["title","genres","predicted_rates"]].nlargest(n_top,"predicted_rates")
```

# Create a function

```python
def get_recommendation ( n=10, user_id= 1):
    chosen_user = user_id
    n_top = n
    cross_table = pd.pivot_table(data=ratings_df, values="rating",columns="movieId",index="userId")
    cross_table = cross_table.fillna(0)
    
    usrs_similarities = pd.DataFrame(cosine_similarity(cross_table),columns=cross_table.index, index=cross_table.index)
    
    user_weight_col = usrs_similarities.query("userId!=@chosen_user")[chosen_user]
    weights = user_weight_col/sum(user_weight_col)
    
    not_rated_items = cross_table.loc[cross_table.index!=chosen_user , cross_table.loc[chosen_user,:]== 0 ]
    predicted_rates = pd.DataFrame(not_rated_items.T.dot(weights),columns=["predicted_rates"])
    
    recommended_items = predicted_rates.merge(movies_df, on="movieId")[["title","genres","predicted_rates"]].nlargest(n_top,"predicted_rates")
    
    return recommended_items
    
    
```
